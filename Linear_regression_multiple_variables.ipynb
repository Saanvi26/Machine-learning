```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv("Data.csv")
df
```

```python
import math
median_bedroom=math.floor(df.bedroom.median())#Replace NaN with median of the bedroom
median_bedroom
```

```python
df.bedroom=df.bedroom.fillna(median_bedroom)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df[['price']]) #(independent,dependent)
reg.predict([[3000,3,40]])
```

```python
reg.intercept_
```
